# 2.2 EDA

This purpose of this notebook is simply to look at changes in student population over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/interim/all_data.csv')
```

```python
sns.lineplot(data=df, x='year', y='total', hue='school_id')
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=df, x='year', y='total', ax=ax)
ax.set_title('Mean Number of Students Over time')
ax.set_xticks(ticks=(2010, 2011, 2012))
plt.show()
fig.savefig('../reports/figures/students_over_time.png')
```
